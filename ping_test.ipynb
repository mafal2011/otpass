```python
import subprocess

def ping_test(host, timeout=5):
    platform = subprocess.getoutput('uname')  # 현재 OS 확인

    if 'Linux' in platform:  # Linux 경우
        cmd = ['ping', '-c', '4', host]
    elif 'Darwin' in platform:  # MacOS 경우
        cmd = ['ping', '-c', '4', host]
    else:  # Windows 경우
        cmd = ['ping', '-n', '4', host]
    
    try:
        result = subprocess.run(cmd, timeout=timeout)
        return result.returncode == 0  # 성공하면 True, 실패하면 False 반환
    except subprocess.TimeoutExpired:
        return 'timeout'  # 타임아웃 발생 시 문자열 반환
```

```python
def parallel_ping_test(hosts, timeout=5, max_workers=10):
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        results = list(executor.map(ping_test, hosts, [timeout]*len(hosts)))
    return results
```

```python
host = [f'192.168.190.{i}' for i in range(256)]
results = parallel_ping_test(host)
```

```python
for idx, result in enumerate(results):
    if result != 'timeout':
        print(f'192.168.190.{idx} ({result})')
```

```python
host = '8.8.8.8'
if ping_test(host):
    print(f'{host} is reachable.')
else:
    print(f'{host} is not reachable.')
```
